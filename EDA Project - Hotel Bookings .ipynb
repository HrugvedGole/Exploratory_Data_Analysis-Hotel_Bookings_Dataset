## What I learnt about the dataset

```python
# Filling children column with 0 assuming the customer did'nt have children
# Filling agent & company column with 0  assuming the customer did'nt book through them.

df["children"].fillna(0, inplace=True)
df["agent"].fillna(0, inplace=True)
```

```python
# filling the country column with 'other' assuming the country was not mentioned at the time of booking

df["country"].fillna("Other", inplace=True)
```

```python
df.isna().sum()
```

### Dropping the unnecessary data

```python
# Dropping the company column as it contains a lot of null values as compared to the rest of the data
df.drop("company", axis=1, inplace=True)
```

```python
# Dropping those rows where there are no adults, children and babies are available.
no_guest = df[df["adults"] + df["children"] + df["babies"] == 0]
df.drop(no_guest.index, axis=0, inplace=True)
```

### Adding New columns to make the data ready for analysis

```python
# Adding a column to check the total number of guests by adding the number of adults, children and babies.
df["total_guests"] = df["adults"] + df["children"] + df["babies"]

# Adding a column to check the total stay by guests by adding the stays in week nights and stays in weekend nights.
df["total_stay"] = df["stays_in_week_nights"] + df["stays_in_weekend_nights"]
```

## Bivariate and Multivariate Analysis

### 15. How long do people stay in the hotel?

```python
# Creating a not cancelled dataframe
not_cancelled_df = df[df["is_canceled"] == 0]

# Creating a Hotel Stay dataframe which has data about stay till 15 days
hotel_stay_df = not_cancelled_df[not_cancelled_df["total_stay"] <= 15] 

# Visualising
plt.figure(figsize=(15,5))
plt.title("Total Days of Stay by guests")

total_stay_countplot = sns.countplot(hotel_stay_df, x="total_stay", hue="hotel") 
```

##### As we can see, in city hotel,  most guests prefer to stay for 3 days and in resort hotel, most people stay for only 1 day.

### 16. Which Hotel makes the most revenue?

```python
# Finding the revenue for each hotel by using the group by function.
revenue = list(df.groupby("hotel")["adr"].sum())

plt.pie(revenue, labels=["City Hotel","Resort Hotel"], autopct="%.2f%%")
plt.title("Total Revenue of each of the Hotel")
plt.show()
```

##### We can conclude that the City Hotel has more share in the total revenue over the Resort Hotel.

### 17. Which Hotel has the longest wating time?

```python
# Creating a dataframe with the average waiting time of each hotel
avg_waiting_df = df.groupby("hotel")["days_in_waiting_list"].mean().reset_index()

plt.figure(figsize=(7,5))
plt.title("Average waiting time for each hotel")

waiting_barplot = sns.barplot(avg_waiting_df, x="days_in_waiting_list", y="hotel")
```

##### From the above insight we can see that the City Hotel has more waiting time as compared to the Resort Hotel.

### 18. Which Hotel has the most repeated guests?

```python
repeated = df[df["is_repeated_guest"] == 1]
repeated_guests_df = repeated.groupby("hotel")["is_repeated_guest"].count().reset_index()

plt.figure(figsize=(7,5))
plt.title("Repeated Guests by Hotel Type")

repeated_barplot = sns.barplot(repeated_guests_df, x="is_repeated_guest", y="hotel")
```

##### As we can see, the Resort Hotel has slightly more repeated guests than the City Hotel.

### 19.  What is the average daily revenue (adr) across different months?

```python
# Grouping the arrival date month and hotel column on mean of adr 
booking_months_df = df.groupby(["arrival_date_month","hotel"])["adr"].mean().reset_index()

# Creating a month list to order the months in ascending
months = ["January","February","March","April","May","June","July","August","September","October","November","December"]

plt.figure(figsize=(15,7))
plt.title("Average Daily Revenue across different Months for both the hotels")

adr_months_barplot = sns.barplot(booking_months_df, y="adr", x="arrival_date_month", hue="hotel", order=months)
```

##### From the above insight, we can see that for Resort Hotel August month had the highest adr and for City Hotel may month had the highest adr.

### 20. Which distribution channel has the highest average daily revenue?

```python
# Grouping the arrival distribution channel column on mean of adr 
dist_channel_df = df.groupby("distribution_channel")["adr"].mean().reset_index()

plt.figure(figsize=(7,5))
plt.title("Average Daily Revenue across different distribution channels")

adr_months_barplot = sns.barplot(dist_channel_df, x="adr", y="distribution_channel")
```

##### As we can see, GDS channel has the highest average daily revenue.
